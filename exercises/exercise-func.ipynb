## Euclidiean distance
##  (10, 3), (-1, -9), (10, -10), (4, -2), (9, -10). (*)

```python
def euclidian(x1, y1, x2, y2):
    return ((x1-x2)**2 +(y1-y2)**2)**0.5

print(f"{euclidian(10, 3, 0, 0)}")
```

## 2. Mathematical functions
## Make the following functions with def or lambda and plot their graphs in the same figure window, with 
## x [-10 to 10]
 ## y= x**2 -3
 ## y= 4x -7


```python
import matplotlib.pyplot as plt

def f1(x):
    return x**2-3

def f2(x):
    return 4*x-7

xax=[x for x in range(-10, 10)]

y1=[f1(x1) for x1 in xax]
y2=[f2(x2) for x2 in xax]

plt.plot(xax,y1)
plt.plot(xax,y2)
```
